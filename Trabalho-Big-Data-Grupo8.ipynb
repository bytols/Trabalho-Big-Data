**Criando médias dos dados**

```python
sales_2017 = df2['Total Orders'][df2['Year'] == 2017]
sales_2018 = df2['Total Orders'][df2['Year'] == 2018]
sales_2019 = df2['Total Orders'][df2['Year'] == 2019]


mean_sales_2017 = df2['Total Orders'][df2['Year'] == 2017].mean()
mean_sales_2018 = df2['Total Orders'][df2['Year'] == 2018].mean()
mean_sales_2019 = df2['Total Orders'][df2['Year'] == 2019].mean()

mean_sales_price_2017 = df2['Total Sales'][df2['Year'] == 2017].mean()
mean_sales_price_2018 = df2['Total Sales'][df2['Year'] == 2018].mean()
mean_sales_price_2019 = df2['Total Sales'][df2['Year'] == 2019].mean()



print('média de produtos vendidos de 2017: {} / , médias de 2018: {} / e médias de 2019: {}'.format(mean_sales_2017,mean_sales_2018,mean_sales_2019))


print('média de valor de vendas de 2017: {} , / médias de 2018: {} / e médias de 2019: {}'.format(mean_sales_price_2017,mean_sales_price_2018,mean_sales_price_2019))
```

**Criação do indice de Performance de vendas da loja**

```python
media_produtos= (mean_sales_2017 + mean_sales_2018 + mean_sales_2019) / 3 

media_rendimentos = (mean_sales_price_2017 + mean_sales_price_2018 + mean_sales_price_2019) / 3 

print('está é a media da venda dos produtos {} e essa é a dos preço bruto {}'.format(media_produtos,media_rendimentos))
```

```python
mean_sales_2017 = mean_sales_price_2017 * 100  / media_rendimentos

print(mean_sales_2017)
```

**Criando uma nova coluna no dataframe com os indicadores de performance**

```python
df2['indicador Performance'] = df2['Net Sales'].apply(lambda x: x * 100/ media_rendimentos )
df2.head()
```

**Criando um gráfico que mostra o indicador de performance ao longo do tempo da empresa**

- *assim é possivel analisar como a empresa se comportou ao longo do tempo usando o indicador*

```python
fig2 = px.line(df2, y="indicador Performance", x="Mes Ano")
fig2.show()
```
